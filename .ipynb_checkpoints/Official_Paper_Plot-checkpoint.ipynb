```python
def Hilbert_psd_arr(subj_arr, mode = "power"):
    hbert = Hilbert(subj_arr)
    if mode == "phase":
        return(np.angle(hbert))
    elif mode == "power":
        return(abs(hbert))
            
def _find_powers(raw_dat, freq_range, sfreq = 512):
    """
    Returns the connectivity array for all possible pairs of channels in raw_dat over the freq   range specified.
    Connectivity is calculated with hilbert transform every WINDOW_SIZE
    """
    concat_arr = None
    for freqs in freq_range:
        filtered_dat = mne.filter.filter_data(raw_dat, sfreq, freqs[0], freqs[1], filter_length='auto', method='fir', phase='zero-double', fir_window='hamming')
        hbert_power = np.expand_dims(Hilbert_psd_arr(filtered_dat, mode='power'),axis=0)
        if concat_arr is None:
            concat_arr = hbert_power
        else:
            concat_arr = np.concatenate([concat_arr,hbert_power],axis=0)
    return(concat_arr)
```

```python
import pickle

def calculate_powers(subj_name, def_freqs, region='audi'):
    # Process the imported data one by one. Decomposing morlet results into frequency bands
    # 1. freqs = [(1,2),(2,5),(5,14),(14,30),(30,70),(70,150)] 2. BBD
    read_path = "F:\\vmpfc_project\\data\\morlet_128Hz\\"
    #subject_number = 's2'
    #s7_audi_morlet_128.pickle
    with open(read_path+subj_name+"_"+region+"_morlet_128.pickle", 'rb') as ff:
        subject_audi_morlet = pickle.load(ff)
    subject_audi_morlet = abs(subject_audi_morlet)
    #return(subject_audi_morlet)
    subj_power_band = []
    freqs = np.logspace(*np.log10([1, 150]), num=30)

    for ffrq in def_freqs:
        band_a = np.logical_and(np.logical_and(freqs>=ffrq[0], freqs<ffrq[1]), (~ np.logical_and(freqs>56, freqs<64)), (~ np.logical_and(freqs>116, freqs<124)))
        subj_dat_a = subject_audi_morlet[0,:,band_a,:]
        subj_dat_a = np.swapaxes(subj_dat_a,0,1)
        subj_dat_b = np.zeros(subj_dat_a.shape)
        for i in range(subj_dat_b.shape[0]):
            for j in range(subj_dat_b.shape[1]):
                subj_dat_b[i,j,:] = subj_dat_a[i,j,:]/np.mean(subj_dat_a[i,j,:])
        #subj_dat_b = (subj_dat_a / np.mean(subj_dat_a,axis=1,keepdims=True)) * 100 #demean it
        subj_power_band.append(np.mean(subj_dat_b,axis=1))
        #subj_power_band.append(subj_dat_a)
    return subj_power_band

def_freqs = [(1,2),(2,5),(5,14),(14,30),(30,70),(70,151),(1,151)]  

s2_audi_pbd = calculate_powers(subj_name='s2',def_freqs=def_freqs, region='audi')
s4_audi_pbd = calculate_powers(subj_name='s4',def_freqs=def_freqs, region='audi')
s5_audi_pbd = calculate_powers(subj_name='s5',def_freqs=def_freqs, region='audi')
s7_audi_pbd = calculate_powers(subj_name='s7',def_freqs=def_freqs, region='audi')
s10_audi_pbd = calculate_powers(subj_name='s10',def_freqs=def_freqs, region='audi')
s14_audi_pbd = calculate_powers(subj_name='s14',def_freqs=def_freqs, region='audi')

s2_vmpfc_pbd = calculate_powers(subj_name='s2',def_freqs=def_freqs, region='vmpfc')
s4_vmpfc_pbd = calculate_powers(subj_name='s4',def_freqs=def_freqs, region='vmpfc')
s5_vmpfc_pbd = calculate_powers(subj_name='s5',def_freqs=def_freqs, region='vmpfc')
s7_vmpfc_pbd = calculate_powers(subj_name='s7',def_freqs=def_freqs, region='vmpfc')
s10_vmpfc_pbd = calculate_powers(subj_name='s10',def_freqs=def_freqs, region='vmpfc')
s14_vmpfc_pbd = calculate_powers(subj_name='s14',def_freqs=def_freqs, region='vmpfc')
```

```python
# I need to bootstrap and permute raw ISC
import sklearn
def _quickbootstrap(list_A, n_perm = 2000, num_boot = None):
    if num_boot is None:
        num_boot = int(1 * len(list_A))
    boot_mean = []
    for _ in range(n_perm):
        boot = sklearn.utils.resample(list_A, replace=True, n_samples=num_boot)
        boot_avg = np.tanh(np.mean(np.arctanh(boot)))
        boot_mean.append(boot_avg)
    return(boot_mean)

def calculate_pval_rawISC(data_matrix):
    corr_matrix = 1 - pairwise_distances(data_matrix, metric='correlation')
    lower_index = np.tril_indices(corr_matrix.shape[0],-1)
    corr_arr = corr_matrix[lower_index]
    bootstrapped_mean = _quickbootstrap(corr_arr,n_perm=2000)
    return(np.percentile(bootstrapped_mean,5),np.percentile(bootstrapped_mean,100))
    
```

## Steps to calculate broadband power
1. divide recording session to 500 ms epoch (256 data points)
2. Use wavelet transform to calculate broadband power spectrum for each epoch
3. Fit Robust regression of power vs spectrum within each epoch
4. Take the mean height of the robust regression fitted line in each epoch (mean of all the fitted values)

```python
def _modified_compute_isc(data, audi_cumu_idx, return_list = False):
    ''' Helper function to compute intersubject correlation from observations by a nested subjects array, separated by
        within-subject and cross-subject 
        Args:
            data: (pd.DataFrame, np.array) observations by subjects where isc is computed across subjects
            audi_cumu_idx: (np.array) the cumulative index indicating the electrodes
        
        Returns:
            within_isc: (float) withinsubject correlation coefficient
            cross_isc: (float) cross subject correlation coefficient
    '''    
    within_subject_corrs = []
    cross_subject_corrs = []
    
    corr_matrix = 1 - pairwise_distances(data, metric='correlation')
    lower_index = np.tril_indices(corr_matrix.shape[0],-1)
    for i in range(len(lower_index[0])):
        #print(lower_index[0][i])
        immediate_above_idx = next(x[0] for x in enumerate(audi_cumu_idx) if x[1] > lower_index[0][i])
        immediate_below_idx = immediate_above_idx-1
        #print("index aboveL",audi_cumu_idx[immediate_above_idx])
        #print('index below',audi_cumu_idx[immediate_below_idx])
        if (lower_index[1][i] >= audi_cumu_idx[immediate_below_idx]) and (lower_index[1][i] < audi_cumu_idx[immediate_above_idx]):
            #print("within:",lower_index[0][i],lower_index[1][i])
            within_subject_corrs.append(corr_matrix[lower_index[0][i],lower_index[1][i]])
        else:
            #print('cross:',lower_index[0][i],lower_index[1][i])
            cross_subject_corrs.append(corr_matrix[lower_index[0][i],lower_index[1][i]])
    
    within_isc = np.tanh(np.mean(np.arctanh(within_subject_corrs)))
    cross_isc = np.tanh(np.mean(np.arctanh(cross_subject_corrs)))
    if return_list:
        return (within_subject_corrs,cross_subject_corrs)
    else:
        return (within_isc,cross_isc)
```

## All kinds of p value calculation

```python
def _bootstrap_isc(
    similarity_matrix, metric="median", exclude_self_corr=True, random_state=None
):
    """Helper function to compute bootstrapped ISC from Adjacency Instance
    This function implements the subject-wise bootstrap method discussed in Chen et al., 2016.
    Chen, G., Shin, Y. W., Taylor, P. A., Glen, D. R., Reynolds, R. C., Israel, R. B.,
    & Cox, R. W. (2016). Untangling the relatedness among correlations, part I:
    nonparametric approaches to inter-subject correlation analysis at the group level.
    NeuroImage, 142, 248-259.
    Args:
        similarity_matrix: (Adjacency) Adjacency matrix of pairwise correlation values
        metric: (str) type of summary statistic (Default: median)
        exclude_self_corr: (bool) set correlations with random draws of same subject to NaN (Default: True)
        random_state: random_state instance for permutation
    Returns:
        isc: summary statistic of bootstrapped similarity matrix
    """
    from nltools.data import Adjacency

    if not isinstance(similarity_matrix, Adjacency):
        raise ValueError("similarity_matrix must be an Adjacency instance.")

    random_state = check_random_state(random_state)

    square = similarity_matrix.squareform()
    n_sub = square.shape[0]
    np.fill_diagonal(square, 1)

    bootstrap_subject = sorted(
        random_state.choice(np.arange(n_sub), size=n_sub, replace=True)
    )
    bootstrap_sample = Adjacency(
        square[bootstrap_subject, :][:, bootstrap_subject], matrix_type="similarity"
    )

    if exclude_self_corr:
        bootstrap_sample.data[bootstrap_sample.data == 1] = np.nan

    if metric == "mean":
        return np.tanh(bootstrap_sample.r_to_z().mean())
    elif metric == "median":
        return bootstrap_sample.median()


def _compute_isc(data, metric="median"):
    """Helper function to compute intersubject correlation from observations by subjects array.
    Args:
        data: (pd.DataFrame, np.array) observations by subjects where isc is computed across subjects
        metric: (str) type of association metric ['spearman','pearson','kendall']
    Returns:
        isc: (float) intersubject correlation coefficient
    """

    from nltools.data import Adjacency

    similarity = Adjacency(
        data, matrix_type="similarity"
    )
    if metric == "mean":
        isc = np.tanh(similarity.r_to_z().mean())
    elif metric == "median":
        isc = similarity.median()
    return isc


def modified_isc(
    data,
    n_bootstraps=5000,
    metric="mean",
    method="bootstrap",
    ci_percentile=95,
    exclude_self_corr=True,
    return_bootstraps=False,
    tail=2,
    n_jobs=-1,
    random_state=None,
):

    from nltools.data import Adjacency

    random_state = check_random_state(random_state)

    if not isinstance(data, (pd.DataFrame, np.ndarray)):
        raise ValueError("data must be a pandas dataframe or numpy array")

    if metric not in ["mean", "median"]:
        raise ValueError("metric must be ['mean', 'median']")

    stats = {"isc": _compute_isc(data, metric=metric)}

    similarity = Adjacency(
        data, matrix_type="similarity"
    )

    if method == "bootstrap":
        all_bootstraps = Parallel(n_jobs=n_jobs)(
            delayed(_bootstrap_isc)(
                similarity,
                metric=metric,
                exclude_self_corr=exclude_self_corr,
                random_state=random_state,
            )
            for _ in range(n_bootstraps)
        )
        stats["p"] = _calc_pvalue(all_bootstraps - stats["isc"], stats["isc"], tail)

    stats["ci"] = (
        np.percentile(np.array(all_bootstraps), (100 - ci_percentile) / 2, axis=0),
        np.percentile(
            np.array(all_bootstraps), ci_percentile + (100 - ci_percentile) / 2, axis=0
        ),
    )

    if return_bootstraps:
        stats["null_distribution"] = all_bootstraps

    return stats

def _calc_pvalue(all_p, stat, tail):
    """Calculates p value based on distribution of correlations
    This function is called by the permutation functions
        all_p: list of correlation values from permutation
        stat: actual value being tested, i.e., stats['correlation'] or stats['mean']
        tail: (int) either 2 or 1 for two-tailed p-value or one-tailed
    """

    denom = float(len(all_p)) + 1
    if tail == 1:
        numer = np.sum(all_p >= stat) + 1 if stat >= 0 else np.sum(all_p <= stat) + 1
    elif tail == 2:
        numer = np.sum(np.abs(all_p) >= np.abs(stat)) + 1
    else:
        raise ValueError("tail must be either 1 or 2")
    return numer / denom
```

```python
import sklearn
from joblib import Parallel, delayed


def permute_contrats_audivmpfc(auditory_corrs, vmpfc_corrs, nperm=500):
    """
    The function serves to permute and contrast audi vs vmpfc ISC.
    The algorithm will be to permute with replacement the auditory electrodes
    and count the number of times audiory average is larger than vmpfc average 
    """
    sign_flips=[]
    for i in range(nperm):
        audi_boots = sklearn.utils.resample(auditory_corrs, replace=True, n_samples=int(5*len(auditory_corrs)/6))
        vmpfc_boots = sklearn.utils.resample(vmpfc_corrs, replace=True, n_samples=int(5*len(vmpfc_corrs)/6))
        audi_boot_isc = np.tanh(np.mean(np.arctanh(audi_boots)))
        #print(audi_boot_isc)
        vmpfc_boot_isc = np.tanh(np.mean(np.arctanh(vmpfc_boots)))
        #print("audi:",audi_boot_isc)
        #print("vmpfc:",vmpfc_boot_isc)
        sign_flips.append(audi_boot_isc<vmpfc_boot_isc)
    return sign_flips
    
def circle_shift(data, random_state=None):
    '''Circle shift data for each feature
        
    Args:
    
        data: time series (1D or 2D). If 2D, then must be observations by features
        random_state: (int, None, or np.random.RandomState) Initial random seed (default: None)
            
    Returns:
    
        shifted data
    
    '''
    random_state = check_random_state(random_state)
    data = np.array(data)
    if len(data.shape) == 1:
        shift = random_state.choice(np.arange(len(data)), replace=False)
        shifted = np.concatenate((data[-shift:], data[:-shift]))
    else:
        shift = random_state.choice(np.arange(data.shape[0]), size=data.shape[1], replace=False)
        shifted = np.array([np.concatenate([data[-int(s):, int(d)], data[:-int(s), int(d)]]) for d,s in zip(range(data.shape[1]), shift)]).T
    return shifted

def permute_group_compare(group_A, group_B, num_swap = 100, n_iters = 1500):
    """
    Permutation test to compare group differences
    
    """
    #import random
    #random.seed(1)
    perm_mean = []
    group_label = np.repeat([1,2],[len(group_A), len(group_B)])
    all_group_cbd = np.concatenate([group_A,group_B])
    
    for _ in range(n_iters):
        np.random.shuffle(group_label)
        new_listA = [all_group_cbd[i] for i in range(len(group_label)) if group_label[i] == 1]
        new_listB = [all_group_cbd[i] for i in range(len(group_label)) if group_label[i] == 2]
        perm_mean.append(np.mean(new_listA) - np.mean(new_listB))
        
    return perm_mean

def n_sample_bootstrap(samples_list, n_iters):
    
    samples_mean = np.mean(samples_list)
    replications = np.array([np.mean(np.random.choice(samples_list, int(len(samples_list)/2), replace = True)) for _ in range(n_iters)])
    
    print(np.sum(replications<0)/n_iters)
```

```python
import sklearn
def _quickbootstrap(list_A, n_perm = 2000, num_boot = None):
    if num_boot is None:
        num_boot = int(1 * len(list_A))
    boot_mean = []
    for _ in range(n_perm):
        boot = sklearn.utils.resample(list_A, replace=True, n_samples=num_boot)
        boot_avg = np.tanh(np.mean(np.arctanh(boot)))
        boot_mean.append(boot_avg)
    return(boot_mean)
```

```python
def _bootstrap_isc(
    similarity_matrix, metric="median", exclude_self_corr=True, random_state=None
):
    """Helper function to compute bootstrapped ISC from Adjacency Instance
    This function implements the subject-wise bootstrap method discussed in Chen et al., 2016.
    Chen, G., Shin, Y. W., Taylor, P. A., Glen, D. R., Reynolds, R. C., Israel, R. B.,
    & Cox, R. W. (2016). Untangling the relatedness among correlations, part I:
    nonparametric approaches to inter-subject correlation analysis at the group level.
    NeuroImage, 142, 248-259.
    Args:
        similarity_matrix: (Adjacency) Adjacency matrix of pairwise correlation values
        metric: (str) type of summary statistic (Default: median)
        exclude_self_corr: (bool) set correlations with random draws of same subject to NaN (Default: True)
        random_state: random_state instance for permutation
    Returns:
        isc: summary statistic of bootstrapped similarity matrix
    """
    from nltools.data import Adjacency

    if not isinstance(similarity_matrix, Adjacency):
        raise ValueError("similarity_matrix must be an Adjacency instance.")

    random_state = check_random_state(random_state)

    square = similarity_matrix.squareform()
    n_sub = square.shape[0]
    np.fill_diagonal(square, 1)

    bootstrap_subject = sorted(
        random_state.choice(np.arange(n_sub), size=n_sub, replace=True)
    )
    bootstrap_sample = Adjacency(
        square[bootstrap_subject, :][:, bootstrap_subject], matrix_type="similarity"
    )

    if exclude_self_corr:
        bootstrap_sample.data[bootstrap_sample.data == 1] = np.nan

    if metric == "mean":
        return np.tanh(bootstrap_sample.r_to_z().mean())
    elif metric == "median":
        return bootstrap_sample.median()
```

```python
def z_score(arr):
    return((arr-np.mean(arr,axis = 1,keepdims=True))/np.std(arr,axis = 1,keepdims=True))

def _align_locs2eeg(loc_data, loc_names, eeg_chs):
    """
    Aligns the locations with the eeg channels, for future brain plots
    Example usage: _align_locs2eeg(loc_data = s14_total_locations, loc_names = s14_electrode_names, eeg_chs = s14_chs)
    """
    intersection = [x for x in range(len(eeg_chs)) if eeg_chs[x] in frozenset(loc_names)]
    intersection_names = [x for x in eeg_chs if x in frozenset(loc_names)]
    locations = [loc_names.index(i) for i in intersection_names]
    #eeg_selected = [eeg_chs[i] for i in intersection]
    return(intersection,loc_data.iloc[locations,:])

def create_brain_autocorrdecay(electrode_locations, loading_values, mask_size):
            
    data_template = None
    mask_count = np.zeros((mask_size,))
    for i in range(len(electrode_locations)):
        #Create a mask that spans a certain radius around the electrode
        mask_sphere = create_sphere(electrode_locations.iloc[i,:], radius = 2)    
        ma = Brain_Data(mask_sphere)
        mask_count = mask_count+ma.data
        #change the value to the autocorrelation decay values.
        ma2 = np.where(ma.data == 1, loading_values[i] , 0)
        #Store the values into data_template.
        if data_template is None:
            data_template = ma2
        else:
            data_template = data_template+ma2
    #data_template = np.divide(data_template, mask_count, out= np.zeros_like(data_template), where = mask_count!=0)
    return(np.nan_to_num(data_template / mask_count))
```

```python
from nilearn.plotting import plot_stat_map, plot_glass_brain,view_img, view_img_on_surf
from nltools.utils import get_mni_from_img_resolution
from nilearn.plotting import plot_glass_brain
from nltools.prefs import resolve_mni_path

def plot_stats(obj, limit=5, anatomical=None, view='axial', colorbar=False, black_bg=True, draw_cross=False,
    threshold_upper=None, threshold_lower=None, figsize=(15, 2), axes=None, audi_left_msk=None,audi_right_msk=None,v_msk=None, **kwargs):
        """ Create a quick plot of self.data.  Will plot each image separately
        Args:
            limit: (int) max number of images to return
            anatomical: (nifti, str) nifti image or file name to overlay
            view: (str) 'axial' for limit number of axial slices;
                        'glass' for ortho-view glass brain; 'mni' for
                        multi-slice view mni brain; 'full' for both glass and
                        mni views
            threshold_upper: (str/float) threshold if view is 'glass',
                             'mni', or 'full'
            threshold_lower: (str/float)threshold if view is 'glass',
                             'mni', or 'full'
            save: (str/bool): optional string file name or path for saving; only applies if view is 'mni', 'glass', or 'full'.
                            Filenames will appended with the orientation they belong to
        """

        if view == 'axial':
            if threshold_upper is not None or threshold_lower is not None:
                print("threshold is ignored for simple axial plots")
            if anatomical is not None:
                if not isinstance(anatomical, nib.Nifti1Image):
                    if isinstance(anatomical, six.string_types):
                        anatomical = nib.load(anatomical)
                    else:
                        raise ValueError("anatomical is not a nibabel instance")
            else:
                # anatomical = nib.load(resolve_mni_path(MNI_Template)['plot'])
                anatomical = get_mni_from_img_resolution(obj, img_type='plot')

            if obj.data.ndim == 1:
                if axes is None:
                    _, axes = plt.subplots(nrows=1, figsize=figsize)
                plot_01=plot_stat_map(obj.to_nifti(), anatomical,
                              cut_coords=range(-40, 60, 10), display_mode='z',
                              black_bg=black_bg, colorbar=colorbar, draw_cross=draw_cross,
                              axes=axes, **kwargs)
                plot_01.add_contours(audi_left_msk.to_nifti(), levels=[0.5], colors='b')
                plot_01.add_contours(audi_right_msk.to_nifti(), levels=[0.5], colors='b')
                plot_01.add_contours(v_msk.to_nifti(), levels=[0.5], colors='pink')
            else:
                if axes is not None:
                    print("axes is ignored when plotting multiple images")
                n_subs = np.minimum(obj.data.shape[0], limit)
                _, a = plt.subplots(nrows=n_subs, figsize=(figsize[0], len(obj) * figsize[1]))
                for i in range(n_subs):
                    plot_01=plot_stat_map(obj[i].to_nifti(), anatomical,
                                  cut_coords=range(-40, 60, 10),
                                  display_mode='z',
                                  black_bg=black_bg,
                                  colorbar=colorbar,
                                  draw_cross=draw_cross,
                                  axes=a[i],
                                  **kwargs)
                    plot_01.add_contours(audi_left_msk.to_nifti(), levels=[0.5], colors='b')
                    plot_01.add_contours(audi_right_msk.to_nifti(), levels=[0.5], colors='b')
                    plot_01.add_contours(v_msk.to_nifti(), levels=[0.5], colors='pink')
            return
        elif view in ['glass', 'mni', 'full']:
            if obj.data.ndim == 1:
                return plot_brain_manual(obj, how=view, thr_upper=threshold_upper, thr_lower=threshold_lower, audi_left_msk=audi_left_msk,audi_right_msk=audi_right_msk,v_msk=v_msk,**kwargs)
            else:
                raise ValueError("Plotting in 'glass', 'mni', or 'full' views only works with a 3D image")
        else:
            raise ValueError("view must be one of: 'axial', 'glass', 'mni', 'full'.")

def plot_brain_manual(objIn, how="full", thr_upper=None, thr_lower=None, save=False,audi_left_msk=None,audi_right_msk=None,v_msk=None, **kwargs):
    """
    More complete brain plotting of a Brain_Data instance
    Args:
        obj: (Brain_Data) object to plot
        how: (str) whether to plot a glass brain 'glass', 3 view-multi-slice mni 'mni', or both 'full'
        thr_upper: (str/float) thresholding of image. Can be string for percentage, or float for data units (see Brain_Data.threshold()
        thr_lower: (str/float) thresholding of image. Can be string for percentage, or float for data units (see Brain_Data.threshold()
        save (str): if a string file name or path is provided plots will be saved into this directory appended with the orientation they belong to
        kwargs: optionals args to nilearn plot functions (e.g. vmax)
    """
    if thr_upper or thr_lower:
        obj = objIn.threshold(upper=thr_upper, lower=thr_lower)
    else:
        obj = objIn.copy()

    views = ["x", "y", "z"]
    coords = [
        range(-50, 51, 8),
        range(-80, 50, 10),
        range(-40, 71, 9),
    ]  # [-88,-72,-58,-38,-26,8,20,34,46]
    #if 'cmap' not in kwargs:
    #    cmap = "RdBu_r"

    if thr_upper is None and thr_lower is None:
        print("Plotting unthresholded image")
    else:
        if isinstance(thr_upper, str):
            print("Plotting top %s of voxels" % thr_upper)
        elif isinstance(thr_upper, (float, int)):
            print("Plotting voxels with stat value >= %s" % thr_upper)
        if isinstance(thr_lower, str):
            print("Plotting lower %s of voxels" % thr_lower)
        elif isinstance(thr_lower, (float, int)):
            print("Plotting voxels with stat value <= %s" % thr_lower)

    if save:
        path, filename = os.path.split(save)
        filename, extension = filename.split(".")
        glass_save = os.path.join(path, filename + "_glass." + extension)
        x_save = os.path.join(path, filename + "_x." + extension)
        y_save = os.path.join(path, filename + "_y." + extension)
        z_save = os.path.join(path, filename + "_z." + extension)
    else:
        glass_save, x_save, y_save, z_save = None, None, None, None

    saves = [x_save, y_save, z_save]
    
    tk_cmap = matplotlib.colors.ListedColormap(['deeppink', 'brown', 'yellowgreen', 'yellow', 'purple', 'darkgreen','cyan'])
    tk_cmap.set_over('0.25')
    tk_cmap.set_under('0.75')
    boundaries = [1,2,3,4,5,6]
    tk_norm = matplotlib.colors.BoundaryNorm(boundaries, tk_cmap.N)

    if how == "full":
        plot_01=plot_glass_brain(
            obj.to_nifti(),
            display_mode="lzry",
            colorbar=True,
            #cmap=cmap,
            plot_abs=False,
            cmap = tk_cmap,
            norm=tk_norm,
            **kwargs
        )
        plot_01.add_contours(audi_left_msk.to_nifti(), levels=[0.5], colors='b')
        plot_01.add_contours(audi_right_msk.to_nifti(), levels=[0.5], colors='b')
        #plot_01.add_contours(v_msk.to_nifti(), levels=[0.5], colors='pink')
        
        if save:
            plt.savefig(glass_save, bbox_inches="tight",dpi=250)
        for v, c, savefile in zip(views, coords, saves):
            plot_01 = plot_stat_map(
                obj.to_nifti(),
                cut_coords=c,
                display_mode=v,
                cmap = tk_cmap,
                norm=tk_norm,              
                #cmap=cmap,
                bg_img=resolve_mni_path(MNI_Template)["brain"],
                **kwargs
            )
            plot_01.add_contours(audi_left_msk.to_nifti(), levels=[0.5], colors='b')
            plot_01.add_contours(audi_right_msk.to_nifti(), levels=[0.5], colors='b')
            #plot_01.add_contours(v_msk.to_nifti(), levels=[0.5], colors='pink')
            if save:
                plt.savefig(savefile, bbox_inches="tight",dpi=250)
    elif how == "glass":
        plot_01 = plot_glass_brain(
            obj.to_nifti(),
            display_mode="lzry",
            colorbar=True,
            cmap = tk_cmap,
            norm=tk_norm,            
            #cmap=cmap,
            plot_abs=False,
            **kwargs
        )
        plot_01.add_contours(audi_left_msk.to_nifti(), levels=[0.5], colors='b')
        plot_01.add_contours(audi_right_msk.to_nifti(), levels=[0.5], colors='b')
        #plot_01.add_contours(v_msk.to_nifti(), levels=[0.5], colors='pink')
                
        
        if save:
            plt.savefig(glass_save, bbox_inches="tight",dpi=250)
    elif how == "mni":
        for v, c, savefile in zip(views, coords, saves):
            plot_01=plot_stat_map(
                obj.to_nifti(),
                cut_coords=c,
                display_mode=v,
                cmap = tk_cmap,
                norm=tk_norm,               
                #cmap=cmap,
                bg_img=resolve_mni_path(MNI_Template)["brain"],
                **kwargs
            )
            plot_01.add_contours(audi_left_msk.to_nifti(), levels=[0.5], colors='b')
            plot_01.add_contours(audi_right_msk.to_nifti(), levels=[0.5], colors='b')
            #plot_01.add_contours(v_msk.to_nifti(), levels=[0.5], colors='pink')

            if save:
                plt.savefig(savefile, bbox_inches="tight",dpi=250)
    del obj  # save memory
    return 
```

```python
def dice_loss(pred, target, smooth = 1):
    """This definition generalize to real valued pred and target vector.
    This should be differentiable.
    pred: tensor with first dimension as batch
    target: tensor with first dimension as batch
    """

    # have to use contiguous since they may from a torch.view op
    iflat = pred.contiguous().view(-1)
    tflat = target.contiguous().view(-1)
    intersection = (iflat * tflat).sum()

    A_sum = torch.sum(iflat * iflat)
    B_sum = torch.sum(tflat * tflat)

    return 1 - ((2. * intersection + smooth) / (A_sum + B_sum + smooth)) / iflat.size(0)
```
